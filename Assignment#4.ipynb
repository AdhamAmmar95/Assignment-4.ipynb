```python

import networkx as nx
import matplotlib.pyplot as plt

X=nx.Graph()

X.add_node(0,pos=(5,2))
X.add_node(1,pos=(4,2))
X.add_node(2,pos=(3,2))
X.add_node(3,pos=(2,1))
X.add_node(4,pos=(1,0))
X.add_node(5,pos=(0,1))
X.add_node(6,pos=(0,3))
X.add_node(7,pos=(1,4))
X.add_node(8,pos=(2,3))
X.add_node(9,pos=(1,2))

X.add_edges_from([(0,1),(1,2),(2,3),(2,8),(3,4),(3,5),(3,8),(3,9),(4,5),
                  (4,9),(5,6),(5,9),(6,7),(6,8),(6,9),(7,8),(7,9),(8,9)])

pos=nx.get_node_attributes(X,'pos')

nx.draw(X,pos,with_labels=True)

plt.show()
```
